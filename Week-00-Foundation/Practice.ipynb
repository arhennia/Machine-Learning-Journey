Python Data Types

```python
age = 20            #integers
cgpa = 9.78         #float
name = "Arti"       #string
color = "black"
is_student = True   #boolean
is_failed = False
x = 2 + 3j          #complex
```

```python
fruits = ["Apple", "Banana", "Mango"] #Lists (order can be changed)
#access
print(fruits[0])
```

```python
numbers = (10,20,30)            #tuples (order cannot be changed)
colors = {"Red","Blue","Green"} #set (unique values only)
#duplicate values disappear for example
x = {1,2,2,3}
print(x)
```

```python
student = {             #dictionary (Stores data as Key: Value)
    "name":"Arti",
    "age":20,
    "cgpa":9.78
}
print(student["name"])
```
